```python
import sympy
import sympy.physics.mechanics as mech
sympy.init_printing()
sympy.physics.vector.init_vprinting()
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy_utils
import scipy.integrate
import collections
import control
```

```python
plt.rcParams['font.size']=12
plt.rcParams['lines.linewidth']=2
```

```python
M, g, l, theta, t, k_p, k_d, u, d = sympy.symbols('M g l theta t k_p k_d, u, d')
```

```python
frame_i = mech.ReferenceFrame('i')
frame_b = frame_i.orientnew('b', 'Axis', (theta(t), frame_i.z))
point_o = mech.Point('o')
point_o.set_vel(frame_i, 0)
point_cm = point_o.locatenew('cm', l*frame_b.y)
point_cm.set_vel(frame_b, 0)
point_cm.v1pt_theory(point_o, frame_i, frame_b)
ball = mech.Particle('ball', point_cm, M)
ball.set_potential_energy(M*g*l*sympy.cos(theta(t)))
Lagrangian = mech.Lagrangian(frame_i, ball)
```

```python
lm = mech.LagrangesMethod(
    Lagrangian=Lagrangian, q_list=[theta(t)], coneqs=None,
    forcelist=[
        (frame_b, (u(t) + d(t))*frame_i.z),
    ],
    frame=frame_i)
lm.form_lagranges_equations()
```

```python
u_eq = -k_p*theta(t) -k_d*theta(t).diff(t)
```

```python
u_potential = -sympy.integrate(u_eq.subs({theta(t).diff(t):0, theta(t):theta}), theta).subs(theta, theta(t))
u_potential
```

```python
E = ball.potential_energy + ball.kinetic_energy(frame_i) + u_potential
E
```

```python
x_vect = sympy.Matrix([theta(t), theta(t).diff(t)])
x_vect
```

```python
dE_dx = sympy.Matrix([E]).jacobian(x_vect)
dE_dx
```

```python
rhs = lm.rhs()
rhs
```

```python
x_dot = rhs.subs(u(t), u_eq)
x_dot
```

```python
E_dot = dE_dx.dot(x_dot).simplify().expand()
E_dot
```

Now we plot all possible energy growths resuling from u for various states and see that the extremum occur at the boundary.

```python
d_vals = np.linspace(-1,1, 10)
for k_d_val in np.linspace(0, 1, 3):
    for theta_dot_val in np.linspace(-1, 1, 4):
        f_E_dot = sympy.lambdify(d, E_dot.subs(
            {k_d: k_d_val, theta(t).diff(t):theta_dot_val, d(t):d}))
        E_dot_vals = f_E_dot(d_vals)
        plt.plot(d_vals, E_dot_vals, 'r')
        plt.grid()
        plt.ylabel('$\dot{E}$')
        plt.xlabel('d')
```

Since $\frac{\partial}{\partial d}\dot{E}=\dot{\theta}$ and is not dependent on $d$, then we must check the boundaries of the allowed domain $d \in [-m,m]$. Noting that when $\theta < 0$ the optimal value is $d=-m$ and $\theta > 0$ the optimal value is $d=m$ , the optimal strategy to maximize $\dot{E}$ is $d(t) = \text{sgn}(\dot{\theta})m$

```python
def define_functions(const):
    u_vect = sympy.Matrix([u(t), d(t)])
    f_pend, jac_pend = sympy_utils.rhs_to_scipy_ode(rhs, t, x_vect, u_vect, const)
    def f_d(t, x, u):
        theta_d = x[1]
        d = const['d_max']
        if x[1] < 0:
            d = -d
        return d
    def f_d_2(t, x, u):
        return const['d_max']*np.sin(2*np.pi*np.sqrt(const['l']/const['g'])*t)
    f_u = sympy_utils.rhs_to_scipy_ode(sympy.Matrix([-k_p*theta(t) - k_d*theta(t).diff(t)]), t, x_vect, [], const)[0]
    f_E = sympy_utils.rhs_to_scipy_ode(sympy.Matrix([E]), t, x_vect, u_vect, const)[0]
    f_E_dot = sympy_utils.rhs_to_scipy_ode(sympy.Matrix([E_dot]), t, x_vect, u_vect, const)[0]
    return {'f_pend':f_pend, 'jac_pend': jac_pend, 'f_d': f_d, 'f_u': f_u, 'f_E': f_E, 'f_E_dot':f_E_dot, 'f_d_2': f_d_2} 
```

```python
class Data(collections.namedtuple('DataRaw', 't x d u E E_dot')):

    @classmethod
    def allocate(cls, n_t, n_x):
        return cls(
            t=np.zeros(n_t),
            x=np.zeros((n_t, n_x)),
            u=np.zeros(n_t),
            d=np.zeros(n_t),
            E=np.zeros(n_t),
            E_dot=np.zeros(n_t))
        
    def set(self, i, t, x, u, d, E, E_dot):
        self.t[i] = t
        self.x[i,:] = x
        self.u[i] = u
        self.d[i] = d
        self.E[i] = E
        self.E_dot[i] = E_dot
        
    def truncate(self, i):
        return Data(
            t=self.t[:i],
            x=self.x[:i,:],
            u=self.u[:i],
            d=self.d[:i],
            E=self.E[:i],
            E_dot=self.E_dot[:i])
```

```python
def saturate(v, v_max):
    v_norm = np.linalg.norm(v)
    if v_norm > v_max:
        v = v*v_max/v_norm
    return v
```

```python
def sim(tf, x0, funcs, f_d): 
    f_pend = funcs['f_pend']
    jac_pend = funcs['jac_pend']
    f_u = funcs['f_u']
    f_E = funcs['f_E']
    f_E_dot = funcs['f_E_dot']
    ode = scipy.integrate.ode(f_pend, jac_pend)
    ode.set_initial_value(x0)
    dt = 0.01
    n_t = np.floor(tf/dt) + 1
    n_x = len(x0)
    data = Data.allocate(n_t, n_x)
    i = 0
    while ode.t < tf:
        x = ode.y
        t = ode.t
        theta = x[0]
        theta_dot = x[1]
        u = f_u(t, x, [])
        u = saturate(u, const['u_max'])
        d = f_d(t, x, u)
        ode.set_f_params([u, d])
        ode.set_jac_params([u, d])
        ode.integrate(t + dt)
        data.set(i, t, x, u, d, f_E(t, x, u), f_E_dot(t, x, [u,d]))
        i = i + 1
    data = data.truncate(i)
    return data
```

```python
const = {'M':1, 'g':9.8, 'l': 1, 'k_p':20, 'k_d':6, 'd_max': 5.6, 'u_max':10}
funcs = define_functions(const)
data = sim(40, [0,0], funcs, funcs['f_d_2'])
E_dot_num = np.hstack([0, np.diff(data.E)])
h = []
h.append(plt.plot(data.t, data.x[:,0])[0])
h.append(plt.plot(data.t, data.d, '-')[0])
h.append(plt.plot(data.t, data.u, '-')[0])
plt.legend(h, ['x','d','u'], loc='best')
plt.title('sinusoidal disturbance at resonance')
```

```python
data2 = sim(40, [0,0], funcs, funcs['f_d'])
E_dot_num = np.hstack([0, np.diff(data2.E)])
h = []
h.append(plt.plot(data2.t, data2.x[:,0])[0])
h.append(plt.plot(data2.t, data2.d, '-')[0])
h.append(plt.plot(data2.t, data2.u, '-')[0])
plt.legend(h, ['x','d','u'], loc='best')
plt.title('max($\dot{E}$) disturbance')
```

```python
def E_dot_attack(y):
    u_part = -const['k_d']*y**2
    u_part = saturate(u_part, const['u_max'])
    d_part = const['d_max']*np.abs(y)
    return u_part + d_part
```
