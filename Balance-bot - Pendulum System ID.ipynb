```python
with open('data/log_sysid_openloop/sess007/log001.csv') as f:
    log = px4_logutil.px4_log_to_namedtuple(f)
```

```python
def get_data_from_log(log, i_start, i_end, cpr=3200):
    t = (log.TIME.StartTime[i_start:i_end] - log.TIME.StartTime[i_start])/1e6
    dt = savitzky_golay(t, 21, 1, 1)
    
    theta = -log.ATT.Pitch[i_start:i_end]
    theta_dot = log.ATT.PitchRate[i_start:i_end]
    theta_ddot_from_theta_dot = savitzky_golay(theta_dot, 21, 1, 1)/dt
    T = -m*g*l*sin(theta)

    alpha0_dot = log.ENCD.vel0[i_start:i_end]*2*pi/cpr
    alpha1_dot = log.ENCD.vel1[i_start:i_end]*2*pi/cpr
    alpha0 = log.ENCD.cnt0[i_start:i_end]*2*pi/cpr
    alpha1 = log.ENCD.cnt1[i_start:i_end]*2*pi/cpr
    alpha = (alpha0 + alpha1)/2
    alpha_dot = (alpha0_dot + alpha1_dot)/2
    alpha_dot_smooth = savitzky_golay(alpha_dot, 21, 1)
    
    alpha_dot_from_alpha = savitzky_golay(alpha, 21, 1, 1)/dt
    alpha_ddot_from_alpha_dot = savitzky_golay(alpha_dot, 51, 1, 1)/dt
    duty_pwm = (log.OUT1.Out0[i_start:i_end] + log.OUT1.Out1[i_start:i_end])/2
    duty = (duty_pwm - 1500)/1500
    V_batt = log.BATT.V[i_start:i_end]
    V = V_batt*duty
    return px4_logutil.dict_to_namedtuple('data', locals())
```

```python
def plot_torque_velocity(data):
    plt.figure(figsize=(10,10))
    plt.subplot(211)

    plt.plot(data.t, data.T)
    plt.grid()
    plt.ylabel('torque, N-m')

    plt.subplot(212)
    plt.plot(data.t, rad2deg(data.alpha_dot), '.')
    plt.plot(data.t, rad2deg(data.alpha_dot_smooth))
    plt.plot(data.t, rad2deg(data.alpha_dot_from_alpha))
    plt.ylabel(r'$\dot{\alpha}$, deg/s')
    plt.xlabel('t, sec')
    plt.grid()
    plt.legend(['measured', 'smoothed', 'from count'], loc='best')
```

```python
data_fit = get_data_from_log(log, 1050, 3500)
plot_torque_velocity(data_fit)
```

Check that numerical derivative of counts, approximately equls the measured encoder velocity.

## Simulation

```python
def pend_dynamics(t, x, m, g, l, J, k_damp):
    theta = x[0]
    theta_dot = x[1]
    T = -m*g*l*sin(theta)
    return [theta_dot, T/J - 2*theta_dot*k_damp/J]
```

```python
def do_sim(k_damp, J, data, plot=False):
    sim = scipy.integrate.ode(pend_dynamics)
    sim.set_initial_value(y=[data.theta[0], data.theta_dot[0]], t=data.t[0])
    sim.set_f_params(m, g, l, J, k_damp)
    i = 0
    x = np.zeros((len(data.t), 2))
    x[0,:] = sim.y
    for i in range(1,len(data.t)):
        sim.integrate(data.t[i])
        x[i,:] = sim.y
    if plot:
        plt.plot(data.t, rad2deg(data.theta_dot), '.')
        plt.plot(data.t, rad2deg(x[:,1]))
        plt.xlabel('t, sec')
        plt.ylabel(r'$\dot{\theta}$, deg/s')
        plt.legend(['measured', 'estimated'])
        plt.grid()
    r = x[:,1] - data.theta_dot
    return r.dot(r.T)/len(r)
```

## Least Squares Fit

Two formulate this as a least squares problem, we take the numerical derivative of the pitch rate to get the accleration.

```python
X = np.vstack((data_fit.theta_dot, data_fit.theta_ddot_from_theta_dot)).T
Y = data_fit.T
k_damp_lsq, J_tot_lsq = pinv(X).dot(Y)
k_damp_lsq, J_tot_lsq
```

```python
def find_R_sq(f, y):
    r = y - f
    t = y - y.mean(0)
    SS_res = r.dot(r.T)
    SS_tot = t.dot(t.T)
    R_sq = 1 - SS_res/SS_tot
    return R_sq
```

```python
find_R_sq(f=J_tot_lsq*data_fit.theta_ddot_from_theta_dot + k_damp_lsq*data_fit.theta_dot, y=data_fit.T)
```

```python
do_sim(k_damp_lsq, J_tot_lsq, data_fit, plot=True);
savefig('figures/balance-pend-sysid-fit.pdf')
```

## Non-linear Optimization Fit

```python
x_d_opt = scipy.optimize.fmin(lambda xd: do_sim(xd[0], xd[1], data_fit), x0=[k_damp_lsq, J_tot_lsq])
print 'k_damp: {:f} N-m/(rad/s), J_tot: {:f} kg-m^2'.format(*x_d_opt)
```

```python
do_sim(x_d_opt[0], x_d_opt[1], data_fit, plot=True)
savefig('figures/balance-pend-sysid-fit.pdf')
```

Note that at low torque, the motor does not follow the model well. Since the magnitude of this torque disturbance will be small, we can bound it without further complicating the model.

The total moment of inertai found using this fit is composed of the motor inertia, $J_m$, and the pendulum inertia, $J$.

$J_{tot} = 2J_m + J$
