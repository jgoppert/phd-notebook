```python
def get_data_from_log(log, i_start, i_end, cpr=3200):
    smooth_window = 13
    smooth_order = 1
    t = (log.TIME.StartTime[i_start:i_end] - log.TIME.StartTime[i_start])/1e6
    dt = savitzky_golay(t, smooth_window, smooth_order, 1)
    alpha0_dot = log.ENCD.vel0[i_start:i_end]*2*pi/cpr
    alpha1_dot = log.ENCD.vel1[i_start:i_end]*2*pi/cpr
    alpha0 = log.ENCD.cnt0[i_start:i_end]*2*pi/cpr
    alpha1 = log.ENCD.cnt1[i_start:i_end]*2*pi/cpr
    alpha = (alpha0 + alpha1)/2
    alpha_dot = (alpha0_dot + alpha1_dot)/2
    

    alpha_dot_smooth = savitzky_golay(alpha_dot, smooth_window, smooth_order)
    
    alpha_dot_from_alpha = savitzky_golay(alpha, smooth_window, smooth_order, 1)/dt
    alpha_ddot_from_alpha_dot = savitzky_golay(alpha_dot, smooth_window, smooth_order, 1)/dt
    duty_pwm = (log.OUT1.Out0[i_start:i_end] + log.OUT1.Out1[i_start:i_end])/2
    duty = (duty_pwm - 1500)/1500
    V_batt = log.BATT.VFilt[i_start:i_end]
    V = V_batt*duty
    return px4_logutil.dict_to_namedtuple('data', locals())
```

```python
def plot_alpha_dot_data(data):
    plt.figure(figsize=(20,5))
    plt.plot(data.t, data.duty)
    plt.plot(data.t, data.V)
    plt.plot(data.t, data.alpha_dot_smooth)
    plt.legend(['duty', 'V, Volts', r'$\dot{\alpha}$, rad/s'])
    plt.xlabel('t, sec')
    plt.grid()
```

## Simulation

```python
def motor_dynamics(t, x, k_damp_over_J, k_emf_over_J, V):
    alpha = x[0]
    alpha_dot = x[1]
    return [alpha_dot, V*k_emf_over_J - alpha_dot*k_damp_over_J]
```

```python
def do_sim(k_damp, k_emf, delay, data, plot=False):
    V_delay = roll(data.V, delay)
    alpha = data.alpha
    alpha_dot = data.alpha_dot
    x0 = np.array([alpha[0], alpha_dot[0]])
    sim = scipy.integrate.ode(motor_dynamics)
    sim.set_initial_value(y=x0, t=data.t[0])
    i = 0
    x = np.zeros((len(data.t), 2))
    x[0,:] = sim.y
    for i in range(1,len(data.t)):
        sim.set_f_params(k_damp, k_emf, V_delay[i])
        sim.integrate(data.t[i])
        x[i,:] = sim.y
    if plot:
        plt.plot(data.t, rad2deg(alpha_dot).max(0)*data.duty/data.duty.max(0), '-')
        plt.plot(data.t, rad2deg(alpha_dot), '.')
        plt.plot(data.t, rad2deg(x[:,1]))
        plt.xlabel('t, sec')
        plt.ylabel(r'$\theta$, deg/s')
        plt.legend(['scaled duty cycle', r'$\dot{\alpha}$ measured', r'$\dot{\alpha}$ predicted'], loc='best')
        plt.grid()
    r = x[:,1] - data.alpha_dot
    ss = r.dot(r.T)/len(r)
    return ss
```

```python
def lsq_fit(data, delay):
    X = vstack((-data.alpha_dot, roll(data.V, delay))).T
    Y = np.array([data.alpha_ddot_from_alpha_dot]).T
    P = pinv(X).dot(Y)
    YH = X.dot(P)
    r = YH - Y
    ss = r.T.dot(r)[0,0]/len(r)
    return P , ss
```

The differential equation can be written in the form:
\begin{align*}
Y = X \Theta \\
\Theta  = X^+Y
\end{align*}
where
\begin{align}
Y = \begin{bmatrix}
\ddot{\alpha}_0 \\
\ddot{\alpha}_1 \\
\vdots
\end{bmatrix} &
X = \begin{bmatrix}
-\dot{\alpha}_0 & V_0 \\
-\dot{\alpha}_1 & V_1 \\
\vdots & \vdots
\end{bmatrix} &
\Theta = \begin{bmatrix}
k_{damp}/J \\ k_{emf}/J
\end{bmatrix} &
\end{align}

The pseudo-inverse can be then used to obtain the least squares solution of the parameters:
\begin{align*}
\Theta  = (X^T X)^{-1} X^T Y
\end{align*}

```python
I_s, V_s, tau_s, V_nl, I_nl, omega_nl = sympy.symbols('I_s, V_s, tau_s, V_nl, I_nl, omega_nl')
motor_constants = {I_s:5, tau_s:1.2, V_s:12, V_nl:12, I_nl:0.3, omega_nl:200*2*pi/60}
motor_constants
```

Source: http://mplab.ucsd.edu/tutorials/dc.pdf

```python
V, J, L, R, I, k_b, alpha, tau, k_tau, nu, t = \
    sympy.symbols('V, J, L, R, I, k_b, alpha, tau, k_tau, nu, t')
I_s, V_s, tau_s, V_nl, I_nl, omega_nl = sympy.symbols('I_s, V_s, tau_s, V_nl, I_nl, omega_nl')
eom1 = -V(t) + L*I(t).diff(t) + R*I(t) + k_b*alpha(t).diff(t)
eom2 = -J*alpha(t).diff(t,2) + k_tau*I(t) - nu*alpha(t).diff(t) - tau
```

```python
mech.mlatex(sympy.Eq(eom1,0))
```

```python
sympy.Eq(eom2,0)
```

```python
sub_stall = {V(t):V_s, I(t).diff(t):0, I(t): I_s, tau: tau_s,
             alpha(t).diff(t,2): 0, alpha(t).diff(t):0}
sub_no_load = {V(t):V_nl, tau: 0, I(t).diff(t):0, I(t): I_nl,
                alpha(t).diff(t): omega_nl, alpha(t).diff(t,2):0}
```

```python
sol_motor = sympy.solve([eom1.subs(sub_stall), eom2.subs(sub_stall),
                   eom1.subs(sub_no_load), eom2.subs(sub_no_load)], nu, k_tau, k_b, R)
sol_motor
```

```python
motor_constants = {I_s:5, tau_s:1.2, V_s:12, V_nl:12, I_nl:0.3, omega_nl:200*2*pi/60}
motor_constants.update({k_b: sol_motor[k_b].subs(motor_constants),
    k_tau: sol_motor[k_tau].subs(motor_constants),
    nu: sol_motor[nu].subs(motor_constants),
    R: float(sol_motor[R].subs(motor_constants))})
motor_constants
```

```python
sol_motor_dynamics = sympy.solve([eom1, eom2], [I(t), alpha(t).diff(t,2)])
motor_eom = ((alpha(t).diff(t,2) - sol_motor_dynamics[alpha(t).diff(t,2)].\
                      expand().subs(L,0))*J).expand().collect(alpha(t).diff(t))
sympy.Eq(motor_eom)
```

```python
print mech.mlatex(motor_eom)
```

```python
k_damp_spec = motor_eom.diff(alpha(t).diff(t)).subs(motor_constants)
k_damp_spec
```

```python
k_emf_spec = -motor_eom.diff(V(t)).subs(motor_constants)
k_emf_spec
```
