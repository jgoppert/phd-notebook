```python
# setup

# ipython setup
%load_ext autoreload
%autoreload  2
%pdb on
%matplotlib inline

# imports
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
import control
import inv_pend_dynamics
import sympy
from IPython.display import Latex
sympy.init_printing()

# matplotlib setup
plt.rcParams['figure.figsize'] = (15,10)
plt.rcParams['font.size'] = 15
```

```python
# system definition
eq = inv_pend_dynamics.derive()
m = eq['const']['m']
l = eq['const']['l']
A = np.array(eq['A'].subs(eq['const']).subs({'k1':0,'k2':0, 'theta(t)':0})).astype(float)
B = np.array([[0],[1/(m*l)]])
C = np.eye(2)
D = np.zeros((2,1))
sys = control.ss(A, B, C, D)
```

```python
# solve for LQR gains
Q = np.eye(2)
R = 100*np.eye(1)
K = control.lqr(sys.A, sys.B, Q, R)[0]
print('K:', K)
```

```python
# plot for LQR gains
inv_pend_dynamics.compare_lyap(eq, K[0,0], K[0,1], 10, 0.01, 0)
plt.xlabel('$\\theta$, radians')
plt.ylabel('$\dot{\\theta}$, radians/second')
plt.title('LQR controller, K=[{:g}, {:g}]'.format(K[0,0], K[0,1]));
```

```python
# plot for non optimal gains
inv_pend_dynamics.compare_lyap(eq, 10, 2, 10, 0.01, 0)
plt.xlabel('$\\theta$, radians')
plt.ylabel('$\dot{\\theta}$, radians/second')
plt.title('Under-powerd controller, K=[{:g}, {:g}]'.format(10, 2));
```

We see that especially for the case where the gain is low and barely able to overcome the torque from the weight of the pendulum, the non-linear lyapunov function (the mechanical energy), does a better job of bounding the states.

The mechanical energy of the system is given by:

```python
sympy.Eq(sympy.Symbol('E'), eq['E'])
```

And the derivative of this expression is given by:

```python
sympy.Eq(sympy.Symbol('\dot{E}'), eq['E_dot'])
```

Which is always less than zero.
